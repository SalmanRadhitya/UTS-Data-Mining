**MELAKUKAN PEMBERSIHAN DATA**

```python
# Mengisi data yang hilang dengan nilai mean (rata-rata) agar bisa menghilangkan nilai kosong atau negatif seperti NaN
wpvdf['ph'].fillna(wpvdf['ph'].mean(), inplace=True)
wpvdf['Sulfate'].fillna(wpvdf['Sulfate'].mean(), inplace=True)
wpvdf['Trihalomethanes'].fillna(wpvdf['Trihalomethanes'].mean(), inplace=True)
wpvdf = wpvdf.replace('NaN', pd.NA)
wpvdf.dropna(inplace=True)
wpvdf
```

```python
print(wpvdf.isnull().sum())
```

Dapat dilihat baris yang memiliki nilai NaN di kolom penting seperti ph, Sulfate dan Trihalomethanes sudah terhapus,dan nilai-nilai yang tidak standar, seperti nilai negatif atau "NaN" sudah diubah dan dihapus. Hal itu dapat dilihat dengan kode diatas ini


**MELAKUKAN TRANSFORMASI DATA**

Melakukan pengecekan data

```python
wpvdf.dtypes
```

Mengubah Tipe datanya

```python
# Mengubah Tipe datanya
wpvdf['ph'] = wpvdf['ph'].astype('int')
wpvdf['Hardness'] = wpvdf['Hardness'].astype('float64')
wpvdf['Solids'] = wpvdf['Solids'].astype('float64')
wpvdf['Chloramines'] = wpvdf['Chloramines'].astype('float64')
wpvdf['Sulfate'] = wpvdf['Sulfate'].astype('float64')
wpvdf['Conductivity'] = wpvdf['Conductivity'].astype('float64')
wpvdf['Organic_carbon'] = wpvdf['Organic_carbon'].astype('int')
wpvdf['Trihalomethanes'] = wpvdf['Trihalomethanes'].astype('float64')
wpvdf['Turbidity'] = wpvdf['Turbidity'].astype('int')
wpvdf['Potability'] = wpvdf['Potability'].astype('int64')

wpvdf.dtypes
```

```python
wpvdf
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Histogram untuk setiap fitur numerik
for column in wpvdf.columns:
  if wpvdf[column].dtype in ['int64', 'float64']:
    plt.figure(figsize=(8, 6))
    sns.histplot(wpvdf[column], kde=True)
    plt.title(f'Histogram of {column}')
    plt.xlabel(column)
    plt.ylabel('Frequency')
    plt.show()

# Scatter plot untuk melihat hubungan antara dua fitur
plt.figure(figsize=(8, 6))
sns.scatterplot(x='ph', y='Hardness', hue='Potability', data=wpvdf)
plt.title('Relationship between pH and Hardness')
plt.xlabel('pH')
plt.ylabel('Hardness')
plt.show()
```
